```python
import os


def Filemod(path,texttoreplace,texttomodify):
  
    with open(path, "r+") as f:
      l = f.readlines()
        
            # check alphabet by alphabet
      counter = 0
      for i in l:
        if texttoreplace in i:
          # replace valure
          Replacement = i.replace(texttoreplace,texttomodify)

          # changes are made in the list
          l = Replacement
          counter += 1

          # The pre existing text in the file is erased
          f.truncate(0)

          # the modified list is written into
          # the file thereby replacing the old text
          f.writelines(l)
          f.close()
      return"Modified successfully"

        
        
        
 
```

##Demonstrate use of abstract class, multiple inheritance and decorator in python using examples.

```python
#Abstract Class
#An abstract class can be considered as a blueprint for other classes. It allows you to create a set of methods that must be created within any child classes built from the abstract class. A class which contains one or more abstract methods is called an abstract class. An abstract method is a method that has a declaration but does not have an implementation.
# While we are designing large functional units we use an abstract class. When we want to provide a common interface for different implementations of a component, we use an abstract class. 


from abc import ABC, abstractmethod
 
class Polygon(ABC):
 
    @abstractmethod
    def noofsides(self):
        pass
 
class Triangle(Polygon):
 
    # overriding abstract method
    def noofsides(self):
        print("I have 3 sides")
 
class Pentagon(Polygon):
 
    # overriding abstract method
    def noofsides(self):
        print("I have 5 sides")
 
class Hexagon(Polygon):
 
    # overriding abstract method
    def noofsides(self):
        print("I have 6 sides")
 
class Quadrilateral(Polygon):
 
    # overriding abstract method
    def noofsides(self):
        print("I have 4 sides")


class Sedan(Car) :
    def __init__(self, brand,varient, mileage , price , seatingCapacity) -> None:
        super().__init__(brand, varient, mileage)
        self.price = price 
        self.seatingCapacity = seatingCapacity 

class Suv(Sedan):
    def __init__(self, brand,varient, mileage , price , seatingCapacity , groundclearance , tyresize , height) -> None:
        super().__init__(brand,varient, mileage , price , seatingCapacity)
        self.groundclearance = groundclearance
        self.tyresize = tyresize    
        self.height = height
```

```python
#Multiple inheritance
class Car:
    def __init__(self , brand , varient , mileage ) -> None:
        self.brand = brand 
        self.varient = varient 
        self.mileage = mileage

    def specs(self) :
        return f'The brand of car is {self.brand} and the mileage is {self.mileage} \n'

    def typeofvarient(self) :
        if self.price < 1000000 :
            return 'This is a budget car '
        elif self.price < 1000000 : 
          return 'This is a average car '  
        else :
            return 'This is a laxury car '
```

```python
#decorator
def upper_case_decorator(original_function):

    # arg1 is received by the wrapper
    def wrapper(arg1):

        # arg1 is passed to the original function from the wrapper.
        result = original_function(arg1)

        result = result.upper()
        return result

    return wrapper

@upper_case_decorator
def greeting(arg1):
    return arg1 + '!'
```
